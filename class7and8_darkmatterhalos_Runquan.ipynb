We can create a map of all of the small halos that merged to form the larger object - this "map" is called a **merger tree**.

Another reason we care about merger trees. They are critical in order to build *Semi-Analytic Models of galaxy formation (SAMs)*. SAMs are build on top of halo merger trees. 

**NOTE**: the merger trees can either be produced analytically (the details of how to do this are in Frank's lectures note, but we will skip those details here), or built from N-body simulations. ## Halo Bias

We now turn our attention to the spatial distribution of dark matter halos. Dark matter halos form from the dark matter field, so isn't the spatial distribution of dark matter halos the same as the underlying DM distirbution?

The answer is *no* ...

<img width=800 align=left src="slidesfigs/astro610_lecture10-25.png"></img>

There are analytic ways of computing halo bias such as:
* Extended press schecter theory
* Peak background split

**TO REMEMBER**

Using one of these methodologies (the EPS formalism) we can write down that:

$$\delta_h(M,z) = b(M,z) \times \delta(z)$$
 
Where b is the **halo bias** which according to EPS theory is:

$$b(M,z)=1+(\frac{\nu^2-1}{\delta_c})$$

<img width=800 align=left src="slidesfigs/astro610_lecture10-29.png"></img>

**NOTICE**

The mass-bias relation flattens below $M^*$. What does this mean in terms of inferring halo mass from the strength of galaxy clustering for low mass galaxies?

## Assembly bias

<img width=800 align=left src="slidesfigs/astro610_lecture10-31.png"></img>

Assemlbly bias is currently a hot topic in structure formation. Many people are trying to measure this effect with some claimed detections, but nothing robust.

Assembly bias was on of the major themes discussed at this  <a href="https://www.kitp.ucsb.edu/activities/galhalo-c17"> KITP workshop on the Galaxy Halo Connection</a>.

Despite that fact that many of the experts were together in the same room, it was difficult to reach a consensus on the exact *definition* of assembly bias. Some of the main topics to be discussed were:

* assembly bias was originally defined as differences in the clustering of halos at fixed halo mass (e.g., halo spin, halo age). Is this still the definition moving forward (my opinion is yes)?
* does assembly bias manifest itself in galaxy properties?
* how can it be robustly detected - difficulties with cluster finders.

Why is assembly bias now a wide interest? 

* Because data sets are getting more precise and hence our measurements may be now senstitive to this effect. 
* We often use the clustering of galaxies to infer the underlying dark matter halo mass of the sample. If the properties of galaxies correlate with dark matter quanitites other than *mass*, then it may not be true that there is a one to one mapping between the clustering signal and halo mass.

## Halo Mass Definitions

We have seen that the mass of a dark matter halo is generally defined as follows

$$M_{\Delta}=\frac{4}{3} \pi \Delta \rho_{\rm ref} R_{\Delta}^3$$

where:
* $\Delta$ is some overdensity
* $\rho_{\rm ref}$ is either the *mean density* or the *critical density*

However the problem is that:
* there is no "unique" definition of halo mass
* many different halo masses are used throughout the litterature.

Let us review some of the main ones:

* $M_{500c}$: $\Delta=500$, $\rho_{\rm ref}=$critical density
* $M_{200c}$: $\Delta=200$, $\rho_{\rm ref}=$critical density
* $M_{200m}$: $\Delta=200$, $\rho_{\rm ref}=$mean density
* $M_{vir}$: $\Delta=\Delta_{vir}$, $\rho_{\rm ref}=$mean density

NOTE: if you are comparing halo masses, you need to make sure you understand which halo mass is being used, and also how to covert between different halo masses to make sure you are comparing apples to apples. Also, beware of $h$ factors in halo masses and make sure to convert halo masses to your assumed value of $h$.
